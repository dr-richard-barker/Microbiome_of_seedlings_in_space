```python
import requests
import json
import pandas as pd
pd.set_option('display.max_rows', None)
```

```python
# Get all files for OSD-37
response = requests.get('https://osdr.nasa.gov/genelab/data/glds/files/37').json()
```

```python
response
```

Notice in the results above that the API can provide the access to the raw sequencing data "'remote_url': '/geode-py/ws/studies/OSD-37/download?source=datamanager&file=GLDS-37_rna_seq_Atha_Cvi-0_sl-pool_GC_Rep3_R1-GC-D5_R2_raw_fastqc.zip',

```python
# Get the remote URL for all files (need to put https://osdr.nasa.gov before each URL to access)
for item in response['studies']['OSD-37']['study_files']:
    print(item['remote_url'])
```
