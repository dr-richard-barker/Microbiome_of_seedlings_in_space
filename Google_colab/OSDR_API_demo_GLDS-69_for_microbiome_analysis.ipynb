```python
import requests
import json
import pandas as pd
pd.set_option('display.max_rows', None)
```

```python
# Get all files for OSD-69
response = requests.get('https://osdr.nasa.gov/genelab/data/glds/files/69').json()
```

```python
response
```

Notice in the results above that the API can provide the access to the raw sequencing data "'remote_url': '/geode-py/ws/studies/OSD-69/download?


```python
# Get the remote URL for all files (need to put https://osdr.nasa.gov before each URL to access)
for item in response['studies']['OSD-69']['study_files']:
    print(item['remote_url'])
```
